```python
import pandas as pd
import numpy as np
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
boW = CountVectorizer(ngram_range=(1, 1))
tfidf = TfidfVectorizer(ngram_range=(1, 1))
```

```python
df_train = pd.read_csv("cleaned_data.csv")
df_train.head()
```

```python
df_train.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_train.head()
```

```python
X_train = tfidf.fit_transform(df_train["Final_Summary_Reg_Lower_Lem_StopWords_NullRemoval"])
X_train
```

```python
y_train = df_train.Score
y_train.head()
```

```python
df_test = pd.read_csv("cleaned_data_test.csv")
df_test.head()
```

```python
df_test.drop('Unnamed: 0', axis=1, inplace=True)
df_test.head()
```

```python
X_test = tfidf.transform(df_test["Final_Summary_Reg_Lower_Lem_StopWords_NullRemoval"])
X_test
```

```python
y_test = df_test.Score
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier()
```

```python
%%time
# Decision Tree Classifier
dt_classifier.fit(X_train, y_train)

y_pred = dt_classifier.predict(X_test)
y_pred
```

```python
%%time
# RandomForest Classifier
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier()

rf_classifier.fit(X_train, y_train)

y_pred_rf = rf_classifier.predict(X_test)
y_pred_rf
```

```python
y_pred_rf = rf_classifier.predict(X_test)
y_pred_rf
```

```python
%%time
# XGBoost Classifier
from sklearn.ensemble import GradientBoostingClassifier
xgb_classifier = GradientBoostingClassifier()

xgb_classifier.fit(X_train, y_train)

y_pred_xg = xgb_classifier.predict(X_test)
y_pred_xg
```

```python
def metric_class(classifier, y_pred, y_test):
    from sklearn.metrics import accuracy_score, classification_report, precision_score, confusion_matrix
    print("*"*116)
    print("Classifier: ", classifier)
    print("*"*40)
    accuracy_score_ = accuracy_score(y_pred, y_test)
    print("accuracy_score:", accuracy_score_)
    print("*"*40)
#     conf_matrix = confusion_matrix(y_pred, y_test)
#     print("Confusion Matrix:", conf_matrix)
#     print("*"*40)  
    class_report = classification_report(y_pred, y_test)
    print("Classification Report: \n")
    print(class_report)
```

```python
metric_class("Decision Tree classifier", y_pred, y_test)
metric_class("Random Forest classifier", y_pred_rf, y_test)
# metric_class("Decision Tree", y_pred, y_test)
```
